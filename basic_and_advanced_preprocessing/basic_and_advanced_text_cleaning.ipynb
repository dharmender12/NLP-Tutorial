### Handling Emoticons and Emojis
- Emoticons and emojis can convey emotions and sentiments in text data.
- They can be used to enhance sentiment analysis and emotion detection tasks.
- It is important to consider the context in which emoticons and emojis are used, as they can have different meanings in different contexts.
- Some common emoticons include :) for happy, :( for sad, and :D for laughing.
- Emojis can be represented using Unicode characters, and there are various libraries available for handling emojis in text data, such as the `emoji` library in Python.

Ways to handle emoticons and emojis in text data:
1. Remove them: This can be done using regular expressions or by using libraries that specifically target emoticons and emojis.
2. Understand their meaning: This can be done by creating a mapping of emoticons and emojis to their corresponding meanings or sentiments. For example, :) can be mapped to "happy" and :( can be mapped to "sad".

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
data
```

```python
data['data_with_emoji'] = data['data_with_emoji'].apply(remove_emojis)
```

### Text Representation into Vectors
- Text data needs to be converted into a numerical format for machine learning models to process it.
- Common techniques for text representation include:
1. Bag of Words (BoW): This technique represents text as a bag of its words, disregarding grammar and word order but keeping multiplicity. It creates a vector of word counts for each document.
2. Term Frequency-Inverse Document Frequency (TF-IDF): This technique weighs the importance of words in a document based on their frequency in the document and their inverse frequency across all documents. It creates a vector of TF-IDF scores for each word in the document.
3. Word Embeddings: This technique represents words as dense vectors in a continuous vector space. Word embeddings capture semantic relationships between words, allowing for better performance in tasks like sentiment analysis and language modeling. Common word embedding techniques include Word2Vec, GloVe, and FastText.

### Encoding --> 
1. One-Hot Encoding: This technique represents each word as a binary vector, where the length of the vector is equal to the size of the vocabulary. Each position in the vector corresponds to a specific word in the vocabulary, and the value is 1 if the word is present in the document and 0 otherwise.
2. Label Encoding: This technique assigns a unique integer to each word in the vocabulary. The value of the integer corresponds to the index of the word in the vocabulary. This technique is often used for categorical variables, but it can also be applied to text data by treating each word as a category. However, it is important to note that label encoding can introduce ordinal relationships between words, which may not be desirable in certain cases.
3. Ordinal Encoding: This technique is similar to label encoding, but it assigns a unique integer to each word based on its frequency in the corpus. The most frequent word is assigned the integer 1, the second most frequent word is assigned the integer 2, and so on. This technique can be useful for capturing the importance of words in the corpus, but it can also introduce ordinal relationships between words, which may not be desirable in certain cases.
4. Binary Encoding: This technique represents each word as a binary vector, where the length of the vector is equal to the size of the vocabulary. Each position in the vector corresponds to a specific word in the vocabulary, and the value is 1 if the word is present in the document and 0 otherwise. However, unlike one-hot encoding, binary encoding allows for multiple words to be represented in the same vector, which can help reduce the dimensionality of the data.
5. Hashing Encoding: This technique uses a hash function to convert each word into a fixed-length vector. The hash function maps each word to a specific index in the vector, and the value at that index is incremented for each occurrence of the word in the document. This technique can be useful for handling large vocabularies, but it can also lead to collisions, where different words are mapped to the same index in the vector.

#### Properties of Bag of Words
1. In BoW, shapes will remain the same just because every sentence vectorization inclines to the vocabulary itself. 
2. Out of vocabulary words will be ignored in BoW.
3. Semantic Relationship --> BoW does not capture the semantic relationship between words. It treats each word as an independent feature, which can lead to loss of context and meaning in the text data. For example, the sentences "I love AI" and "AI loves me" would have the same BoW representation, even though they convey different meanings.
4. Sparsity --> BoW can lead to sparse vectors, especially when dealing with large vocabularies. This can make it difficult for machine learning models to learn from the data, as many features will have a value of zero.
5. Dimensionality --> BoW can lead to high-dimensional vectors, especially when dealing with large vocabularies. This can make it computationally expensive to process the data and can also lead to overfitting in machine learning models.
6. Ordering of the words --> BoW does not capture the ordering of words in a sentence.

## N-grams and TF-IDF in Text Vectorization

#### One-Hot Encoding : 
- Not even optimal in a way such that we can feed the data to our model. (Shape issue)

#### Bag of Words :
- IT is not having shape issue but it has word sequence is something that has not been taken care of. (Word sequence issue)
- Different meaning sentences can still be considered bit-similar to each other in BoW. (Semantic relationship issue)

#### N-grams (Bi-grams, Tri-grams, etc.) :
- In N-grams, rather than taking a (unique word in the vocabulary), we take the sequence of words in the vocabulary.

- In Bi-grams, (2-consecutive tokens)  are taken together in the vocabulary. 
- In Tri-grams, (3-consecutive tokens) are taken together in the vocabulary.
- In N-grams, (sequential - information) is maintained in the vocabulary.

- For the coding part of (N-grams), we are still going to use countvectorizer --> parameter --> [ngram_range = (1,2)] --> this will take care of both unigrams and bigrams.

## Word2Vec Vector Embeddings in Text Vectorization: 
- Word2Vec is a popular word embedding technique that represents words as dense vectors in a continuous vector space.
- Word2Vec captures semantic relationships between words, allowing for better performance in tasks like sentiment analysis and language modeling.
- Word2Vec can be trained using two main architectures: Continuous Bag of Words (CBOW) and Skip-Gram.
- CBOW predicts the target word based on the context words, while Skip-Gram predicts the context words based on the target word.
- Word2Vec can be implemented using libraries like Gensim in Python, which provides efficient implementations of the Word2Vec algorithm.

#### Word Embeddings:
- Numeric Representation of words in a continuous vector space.
- Dense vectors that capture semantic relationships between words.
- Common techniques include Word2Vec, GloVe, and FastText.
