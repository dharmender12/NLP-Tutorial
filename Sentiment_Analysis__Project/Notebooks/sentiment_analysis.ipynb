### Pipeline for the rating or sentiment analysis of the reviews 
- 4 - 5 stars - Positive
- 3 stars - Neutral
- 1 - 2 stars - Negative

The steps we will follow are:
1. Extract the rating value from the "Rating" column.
2. Create a new column "sentiment" based on the rating value.
3. Analyze the distribution of sentiments in the dataset.
4. lowecase the reviews and remove stop words for further analysis.
5. use TF-IDF vectorization to convert the text data into numerical format for modeling.
6. Train a machine learning model (e.g., Logistic Regression and decision tree) to predict the sentiment based on the reviews.
7. Use SVM to predict the sentiment using SVM and compare the results with the previous models.
8. Evaluate the performance of the models using metrics such as accuracy, precision, recall, and F1-score.



```python
part = df['Review Count'].str.split(' ', expand=True)
# print(part.head())
df['Count'] = part[0].astype('Int32')
# print(df['Review Count'],df['Count'])
```

```python
rating_parts = df['Rating'].astype(str).str.split(" ",expand=True)
# print(rating_parts)
df['rating'] = rating_parts[1].astype('Int32')
df['total_rating'] = rating_parts[4].astype('Int32')
# print(df['rating'],df['total_rating'])
```

```python
df.info()
```

```python
df.drop(columns=['Review Count','Rating'], inplace=True)
```

```python
df.info()
```

```python

df['Review Date'] = pd.to_datetime(df['Review Date'], errors='coerce')
```

### Null value check or missing value check

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['rating'].value_counts()
```

```python
sentiment = []
for i in df['rating']:
    if i >= 4:
        sentiment.append('positive')
    elif i == 3:
        sentiment.append('neutral')
    else:
        sentiment.append('negative')
```

```python
df['sentiment'] = sentiment
```

```python
df['sentiment'].value_counts()
```

```python
df.info()
```

### Encoding 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
## Encoding
onehot = OneHotEncoder()
country_encode = onehot.fit_transform(df[['Country']])
```

```python
le = LabelEncoder()
sentiment_encoded = le.fit_transform(df[['sentiment']])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
df.rename(columns={'Review Text':'review_text'}, inplace=True)
```

```python
df['review_text'] = df['review_text'].str.lower().replace('[^\w\s]','')
```

##  ML Model Building

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
x = tfidf.fit_transform(df['review_text'])
y = sentiment_encoded
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

## Logistic Regression Modelling

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

### Decision Tree Modelling 

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
# print(f"Accuracy of Logistic Regression:\n {classification_report(y_test, y_pred_lr)}")
print(f"Accuracy of Logistic Regression:\n {accuracy_score(y_test, y_pred_lr)}")
```

```python
# print(f"Decision Tree Classification Report:\n {classification_report(y_test, dt.predict(x_test))}")
print(f"Accuracy of Decision Tree:\n {accuracy_score(y_test, dt.predict(x_test))}")
```
